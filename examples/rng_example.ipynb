# Simple random number generator in Python

Python has a built-in module called `random` that allows you to generate random numbers easily. The `numpy.random` module is typically a better option, as it provides more functionality and is faster for large arrays. 

But in computing, there isn't really such a thing as a truly random number. Instead, we use algorithms to generate pseudo-random numbers. These algorithms take an initial value, called a seed, and produce a sequence of numbers that appear random.

Here's a simple algorithm known as the linear congruential random number generator. There are far better algorithms out there, but this one is easy to understand and implement.

The code below uses a set of parameters (a, c, m) to generate a sequence of pseudo-random numbers. The values adopted in this example are commonly used in practice, because they are known to produce good results. If c and m were both even, then the process would generate only even numbers or only odd numbers, which would not be very random.

```python
import matplotlib.pyplot as plt

N=100
a=1664525
c=1013904223
m=2**32

x=1 # seed value
results=[]
for i in range(N):
    x=(a*x+c)%m
    results.append(x)

plt.plot(results,'.')
plt.xlabel('iteration')
plt.ylabel('random number')
```

These numbers look pretty random, but they are not. If you run the code again, you will get the same sequence of numbers. This is because we are using the same seed value. If you want to get a different sequence of numbers, you can change the seed value or use a different algorithm. The fact that we can obtain the same result when we run the algorithm with the same seed means that the algorithm is *deterministic*. This is actually quite useful, because it allows us to reproduce results in scientific calculations. However, it also means that the numbers are not truly random. 

A big problem with this algorithm is that there are correlations between successive values, which can introduce hard-to-detect errors into physics calculations.
