## Probabilities

If an event can happen with some probability *p*, then we can model whether the event happens with the following code.

```python
import random
#random.random() returns a random float in the range [0.0, 1.0]
p=0.2 # event happens 1 out of 5 times on average
def event_happens(p):
    """Returns True if the event happens, False otherwise
    p: probability of the event happening (between 0 and 1)
    """
    chance=random.random()
    # print(chance)
    if chance < p:
        #print("Event happened!")
        return True
    else:
        #print("Event did not happen.")
        return False
```

```python
#test function
happens=0
doesnt_happen=0
for i in range(1000):
    if event_happens(p):
        happens+=1
    else:
        doesnt_happen+=1
print(f"Event happened {happens} times and did not happen {doesnt_happen} times.")
print(f"Empirical probability: {happens/(happens+doesnt_happen)}")
```

Note that as the number of trials increases, the empirical probability approaches *p*. This is known as the law of large numbers.
